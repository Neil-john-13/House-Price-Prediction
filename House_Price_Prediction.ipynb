```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['medv', 'rad'], axis=1)
y = df['medv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
import math

model = LinearRegression()
model.fit(X_train, y_train)

print('Training Score : ', model.score(X_train, y_train))
print('Testing Score  : ', model.score(X_test, y_test))

print('Mean Absolute Error:',mean_absolute_error(y_test,model.predict(X_test)))
print('MSE : ', mean_squared_error(y_test, model.predict(X_test)))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, model.predict(X_test))))
```

```python
print(pd.DataFrame(model.coef_,X.columns,columns=['Coef']))
```

```python

predictions = model.predict(X_test)
plt.scatter(y_test, predictions)
plt.ylabel('Predictions')
plt.title('medv')
```

```python
# Decision tree Regressor

from sklearn.tree import DecisionTreeRegressor
X = df.drop(columns=['medv'], axis=1)
y = df['medv']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)

model1 = DecisionTreeRegressor(max_depth=23, random_state=3)
model1.fit(X_train, y_train)


print('Training Score : ',model1.score(X_train, y_train))
print('Testing Score  : ', model1.score(X_test, y_test))

print('Mean Absolute Error:',mean_absolute_error(y_test,model1.predict(X_test)))
print('MSE : ', mean_squared_error(y_test, model1.predict(X_test)))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, model1.predict(X_test))))
```

```python
predictions = model1.predict(X_test)
plt.scatter(y_test, predictions)
plt.ylabel('Predictions')
plt.title('medv')
plt.show()
```

```python
# Random Forest Regressor

from sklearn.ensemble import RandomForestRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)

model2 = RandomForestRegressor(max_depth = 7, random_state = 63)
model2.fit(X_train, y_train)


print('Training Score : ', model2.score(X_train, y_train))
print('Testing Score  : ', model2.score(X_test, y_test))


```

```python
print('Mean Absolute Error:',mean_absolute_error(y_test,model2.predict(X_test)))
print('MSE : ', mean_squared_error(y_test, model2.predict(X_test)))
print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, model2.predict(X_test))))
```

```python
predictions = model2.predict(X_test)
plt.scatter(y_test, predictions)
plt.ylabel('Predictions')
plt.title('medv')
```
